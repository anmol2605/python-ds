```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
penguins = sns.load_dataset("penguins")
penguins
```

```python
penguins['bill_depth_mm'].plot(linestyle = '', marker = 's', markersize = 10, alpha = 0.5, figsize = (10,5),
                                markerfacecolor = 'yellow',
                                markeredgecolor = 'black'
                                )
```

```python
penguins['bill_depth_mm'][:50].plot(kind= 'bar', figsize = (15,5),color = 'green')
```

```python
penguins['bill_depth_mm'][:50].plot(kind= 'area',alpha = 0.5, figsize = (15,5),color = 'green')
```

```python
print(sns.get_dataset_names())
```

```python
penguins.plot(kind = 'scatter', x = 'bill_length_mm', y = 'bill_depth_mm', s = 100,alpha = 0.5)
```

Subplots

```python
fig, axs = plt.subplots(1,2,figsize = (15,5), constrained_layout = True)
penguins.plot(kind = 'scatter', x = 'bill_length_mm', y = 'bill_depth_mm', s = 100,alpha = 0.5,color = 'red', ax = axs[0], title= ' penguin_bill_data')
penguins.plot(kind = 'scatter', x = 'bill_length_mm', y = 'flipper_length_mm', s = 100,alpha = 0.5, ax = axs[1], title= 'penguin flipper')
plt.show()
```

```python
penguins['species'].value_counts().plot(kind='pie',
                                        autopct = '%1.1f%%',
                                        figsize = (3,3),
                                        title = 'Species',
                                        ylabel='')
```

```python
penguins['species'].value_counts().plot(kind='pie',
                                        autopct = '%1.1f%%',
                                        figsize = (3,3),
                                        title = 'Species',
                                        ylabel='',
                                        pctdistance = .7,
                                        wedgeprops = dict(width = 0.5)
                                        )
```

Advanced visualization


```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.countplot(data=tips, x ='sex')
```

```python
sns.countplot(data = tips, x ='sex', hue='day')
```

```python
sns.displot(data=tips, x='total_bill', binwidth = 3)  #distribution of data
```

```python
sns.displot(data = tips, x = 'total_bill', col= 'sex')
```

```python
sns.displot(data = tips, x = 'total_bill', row= 'sex')
```

```python
sns.displot(data = tips, x = 'total_bill', col = 'smoker', row = 'sex')
```

```python
sns.displot(data = tips, x= 'total_bill', col='smoker', row= 'sex', height = 3, kind = 'kde')
```

```python
sns.displot(data = tips, x= 'total_bill', col='smoker', row= 'sex', height = 3, kind = 'kde', hue='day', multiple = 'stack')
```

```python
sns.displot(penguins, x="flipper_length_mm", hue="species", element="step")
```

```python
sns.displot(penguins, x="flipper_length_mm", kind="kde", bw_adjust=2)
```

```python
sns.displot(penguins, x="bill_length_mm", y="bill_depth_mm")
```

```python
sns.displot(penguins, x="bill_length_mm", y="bill_depth_mm", hue="species")
```

```python
sns.displot(penguins, x="bill_length_mm", y="bill_depth_mm", hue="species", kind="kde")
```

```python
sns.pairplot(penguins)
```

```python
g = sns.PairGrid(penguins)
g.map_upper(sns.histplot)
g.map_lower(sns.kdeplot, fill=True)
g.map_diag(sns.histplot, kde=True)
```

```python
sns.color_palette("Paired")
```

```python
sns.catplot(data=tips, x="day", y="total_bill", jitter=False)
```

```python
sns.catplot(data=tips, x="day", y="total_bill", kind="swarm")
```

```python
sns.catplot(data=tips, x="day", y="total_bill", kind="box")
```

```python
sns.catplot(data=tips, x="day", y="total_bill", hue="smoker", kind="box")
```

```python
diamonds = sns.load_dataset("diamonds")
sns.catplot(
    data=diamonds.sort_values("color"),
    x="color", y="price", kind="boxen",
)
```

```python
sns.catplot(
    data=tips, x="day", y="total_bill", hue="sex",
    kind="violin", split=True,
)
```

```python
sns.catplot(
    data=tips, x="day", y="total_bill", hue="sex",
    kind="violin", inner="stick", split=True, palette="pastel",
)
```

```python
g = sns.catplot(data=tips, x="day", y="total_bill", kind="violin", inner=None)
sns.swarmplot(data=tips, x="day", y="total_bill", color="k", size=3, ax=g.ax)
```
